```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
name="colesterol-1.tif"
#name="ves_full_150_002-1.tif"
Im1 = pylab.imread(name)
fig = figure(1, figsize=(10.5,8.5))
imshow(Im1[:,:],cmap='gray')
plt.title('Imagen Original', fontsize=12)
[xf,yf]=shape(Im1)
```

```python
maxInt=amax(Im1)
minInt=amin(Im1)
Imnew=zeros((xf,yf))
Imnew=((1.0*Im1-minInt)/maxInt)

imshow(Imnew[:,:],cmap='gray')
plt.title('Imagen Original Normalizada', fontsize=12)
```

```python
def loglikelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (-chi_squared)
```

```python
def my_model(x0,y0,r0,r1,m,phi,Im):
    theta=linspace(0,2*pi,10000)
    r=r0+r1*sin(m*theta+phi)
    x=r*cos(theta)+x0
    y=r*sin(theta)+y0
    [xf,yf]=shape(Im)
    Im2=zeros(shape(Im))
    Im2[int_(x), int_(y)] = 1.0
    return Im2
```

```python
imshow(my_model(170,120,40,5,4,pi/2,Im1),cmap='gray')
plt.title('Ejemplo de resultado arrojado por funcion de modelo', fontsize=12)
```

Inicialización de los parámetros. Se inicializó el centro del círculo en el centro de intensidad de la imagen original. m se inicilizó con un entero aleatorio entre 0 y 10, el radio como un número aleatorio normalizado en la mitad del promedio de los límites en x y y de la imagen, r1 como un núero aleatorio de 0 a 1 y phi como un número aleatorio de 0 a 2$\pi$.

```python
x0_walk = empty((0))
y0_walk = empty((0))
r0_walk = empty((0))
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
logl_walk = empty((0))
x0=0
y0=0
for i in range(xf):
    for j in range(yf):
        x0+=Im1[i][j]*i
        y0+=Im1[i][j]*j
x0=x0/float(sum(Im1))
y0=y0/float(sum(Im1))
x0_walk = append(x0_walk, x0)
y0_walk = append(y0_walk, y0)
r0_walk = append(r0_walk, (xf+yf)*random.random()/4.0)
r1_walk = append(r1_walk, random.random())
m_walk = append(m_walk, random.randint(0,5))
phi_walk = append(phi_walk, 2*pi*random.random())

Im_init = my_model(x0_walk[0],y0_walk[0],r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0],Im1)
logl_walk = append(logl_walk, loglikelihood(Imnew, Im_init))
print x0_walk
print y0_walk
print r0_walk
print r1_walk
print m_walk
print phi_walk
print logl_walk
imshow(my_model(x0_walk[0],y0_walk[0],r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0],Im1),cmap='gray')
plt.title('Primera iteracion', fontsize=12)
```

```python
n_iterations = 50000 #this is the number of iterations I want to make
for i in range(n_iterations):
    x0_prime = np.random.normal(x0_walk[i], 1) 
    y0_prime = np.random.normal(y0_walk[i], 1) 
    r0_prime = np.random.normal(r0_walk[i], 0.5) 
    r1_prime = np.random.normal(r1_walk[i], 0.1) 
    m_prime = np.random.randint(0, 5)
    phi_prime = np.random.normal(phi_walk[i], 0.1)
    

    

    Im_init = my_model(x0_walk[i],y0_walk[i],r0_walk[i],r1_walk[i],m_walk[i],phi_walk[i],Im1)
    Im_prime = my_model(x0_prime,y0_prime,r0_prime,r1_prime,m_prime,phi_prime,Im1)
    
    logl_prime = loglikelihood(Imnew, Im_prime)
    logl_init = loglikelihood(Imnew, Im_init)
    
    alpha = exp(logl_prime-logl_init)
    if(alpha>=1.0):
        x0_walk  = append(x0_walk,x0_prime)
        y0_walk  = append(y0_walk,y0_prime)
        r0_walk  = append(r0_walk,r0_prime)
        r1_walk  = append(r1_walk,r1_prime)
        m_walk  = append(m_walk,m_prime)
        phi_walk  = append(phi_walk,phi_prime)
        logl_walk = append(logl_walk, logl_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            x0_walk  = append(x0_walk,x0_prime)
            y0_walk  = append(y0_walk,y0_prime)
            r0_walk  = append(r0_walk,r0_prime)
            r1_walk  = append(r1_walk,r1_prime)
            m_walk  = append(m_walk,m_prime)
            phi_walk  = append(phi_walk,phi_prime)
            logl_walk = append(logl_walk, logl_prime)
        else:
            x0_walk  = append(x0_walk,x0_walk[i])
            y0_walk  = append(y0_walk,y0_walk[i])
            r0_walk  = append(r0_walk,r0_walk[i])
            r1_walk  = append(r1_walk,r1_walk[i])
            m_walk  = append(m_walk,m_walk[i])
            phi_walk  = append(phi_walk,phi_walk[i])
            logl_walk = append(logl_walk, logl_init)
```
